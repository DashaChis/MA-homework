```python
from pymystem3 import Mystem

with open('zh.txt', 'r') as f:
  txt= f.read()[:2000]
print(txt)

m = Mystem()
lemmas = m.lemmatize(txt)

lemmas = ''.join(lemmas)

with open('lemmas.txt', 'w', encoding='utf-8') as f:
  f.write(lemmas)
```

```python
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
from pymorphy2 import MorphAnalyzer
import json

tokens = nltk.word_tokenize(txt)
print(tokens)
morph = MorphAnalyzer()
with open('analiz.jsonl', 'w', encoding='utf-8') as f:
  f.write('')

for w in txt.split():
  print(w)
  ana = morph.parse(w)
  first = ana[0]
  d = {"lemma": first.normal_form, "word": first.word, "pos": first.tag.POS}

  with open('analiz.jsonl', 'a', encoding='utf-8') as f:
    json.dump(d, f, ensure_ascii = False, indent = 4)
```

```python
pos_per = {}

for w in txt.split():
  morph = MorphAnalyzer()
  ana = morph.parse(w)
  first = ana[0]
  if first.tag.POS not in pos_per.keys():
    pos_per[first.tag.POS] = 1
  else:
    pos_per[first.tag.POS] +=1

total = len(txt.split())

for k in pos_per.keys():
  print(k, pos_per[k] / total)
```

```python
verbs = {}
adverbs = {}

for w in txt.split():
  morph = MorphAnalyzer()
  ana = morph.parse(w)
  first = ana[0]
  if 'VERB' in first.tag:
    if first.normal_form not in verbs.keys():
      verbs[first.normal_form] = 1
    else:
      verbs[first.normal_form] += 1
  if 'ADVB' in first.tag:
    if first.normal_form not in adverbs.keys():
      adverbs[first.normal_form] = 1
    else:
      adverbs[first.normal_form] += 1

sorted_tuple = sorted(verbs.items(), key=lambda x: x[1])
for k in sorted_tuple[:20]:
  print(k)

sorted_tuple = sorted(adverbs.items(), key=lambda x: x[1])
for k in sorted_tuple[:20]:
  print(k)
```
